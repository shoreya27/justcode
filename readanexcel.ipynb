```
import xlrd
loc = ("MEDICIN NAMES.xlsx") 
  
wb = xlrd.open_workbook(loc) 
sheet = wb.sheet_by_index(0) 
#sheet.cell_value(0, 0) 
medicine_names_list = list()  
for i in range(sheet.nrows): 
    a = sheet.cell_value(i, 1)
    b = sheet.cell_value(i, 2)
    medicine_names_list.append((a,b))
#print medicine_names_list
medicine_dict_li = list()
for a,b in medicine_names_list:
  d = dict()
  d["medicine"] = str(a)
  d["regimen"] = "M-E-"
  d["duration"] = ""
  d["frequency"] = "daily"
  d["diagnosis"] = ""
  d["formulation"] = str(b)
  d["dosage"] = "1"
  d["remarks"] = ""
  medicine_dict_li.append(d)

print medicine_dict_li

```
